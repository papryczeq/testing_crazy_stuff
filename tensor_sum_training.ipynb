```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
X_train = np.array([
    [1.0, 1.0],
  [1.0, 2.0],
  [2.0, 3.0],
  [3.0, 4.0],
  [4.0, 5.0],
  [6.0, 2.0],
  [7.0, 1.0],
  [8.0, 2.0],
  [9.0, 0.0]
])

y_train = np.array(
    [2.0,
    3.0,
    5.0,
    7.0,
    9.0,
    8.0,
    8.0,
    10.0,
    9.0]
)

X_test = np.array([[5.0, 5.0], [3.0, 1.0], [7.0, 4.0]])
y_test = np.array([10.0, 4.0, 11.0])
```

```python
def create_model():
    model = keras.Sequential([
        keras.layers.Flatten(input_shape=(2,)),
        keras.layers.Dense(1, activation=None, use_bias=False)
    ])
    
    model.compile(optimizer='adam', 
              loss='mse')
    
    return model
```

```python
model = create_model()
```

```python
history = model.fit(X_train, y_train, batch_size=1, epochs=2000)
```

```python
plt.plot(history.history['loss'])
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.evaluate(X_test, y_test)
```
